K-Means Clustering==================Let's say we have data that looks like the following: we have 1000 observationsthat (unknown to us) are random points belonging to five different "classes." One way to think about this is that each datum is generated by a normal distribution around one of five means. The observations of these data each have an x and a y attribute.

```python
observations, real_labels = make_blobs(1000,centers=5, cluster_std=0.7)plt = scatter(observations[:,0], observations[:,1]) 
```

Our mission, then, is to try to derive these groupings using only the attributesthat we've observed (x and y) for each datum.  Minimizing Inertia------------------What K-means does is try to find points in our attribute space (here, a 2-dimensionalspace) that can be thought of as the "centers of gravity" (we'll call them centroids) for the observations we've collected. The assumption is that observations that are members of the same class will gravitate towards each other more than they would towardsmembers of some other class. We're going to start by "guessing" at the number of classes that we have, and then do our best to divide the observations up so that each observation is a member of just oneclass. The goal is for each class to have a stable center of gravity that its points crowd around.If "guessing" in step one strikes you as odd, you're not crazy. The initialization stepof K-Means has an enormous effect on its eventual solution, and there's a whole body ofresearch into how to best accomplish this. Remember that we're guessing at two things:> a. the number of classes that we have (call this K)>> b. our initial stab at where those classes' centroids are locatedIn and of itself, (b) seems like a crazy thing to do. That's okay, though: we're goingto revise our guess (based on actual evidence) often. Guessing at (a), though, is trickier. For demonstration purposes, we'll pick K=5, so we can see K-means succeed. Feel free to try some other values and see how they pan out, though. In practice, wewould try lots of differnt values for K and decide which one gives us the best model of the data.Okay, so we start by randomly picking K points in that space (one for each class):

```python
import numpy as npscatter(observations[:,0], observations[:,1],color='gray') hold(True)# Initialization: Randomly select 5 observations as our centroidsK = 3centroids = np.random.permutation(observations)[:K]scatter(centroids[:,0],centroids[:,1],color='black',marker='x',s=100)centroids
```

and then iterate these steps:> 1. Assign each observation to the class of centroid closest to it> 2. Take the mean of each class, decalre it the new class centroid> 3. Measure the distance that between the old centroid and the new one. >> IF the distance is lower than some threshold, stop.>>>> ELSE continue

```python
guesses = np.zeros_like(real_labels)threshold = 0.1inertia = 100def distance_fct(a,b):    return np.linalg.norm(a - b)def plot_clusters():    """    Just plotting some clusters. Ignore me.    """    scatter(observations[:,0], observations[:,1],             color=map(lambda x: colors[x],guesses))    hold(True)def plot_centroids(centroids,old_centroids):    """    Just plotting some centroids. Ignore me.    """    scatter(centroids[:,0], centroids[:,1],             color='black',s=100,marker='x')    hold(True)    plot([centroids[:,0],old_centroids[:,0]],          [centroids[:,1],old_centroids[:,1]],         color='black')    def step_one():    """     1. Assign each observation to the class of the centroid closest to it    """    for id,observation in enumerate(observations):        distances = [distance_fct(observation,centroid) for centroid in centroids]        distances = np.array(distances)        guesses[id] = distances.argmin()    plot_clusters()    def step_two():    """    2. Take the mean of each class. Assign it as the new centroid for that        class    """    old_centroids = centroids.copy()    for i in range(len(old_centroids)):        centroids[i] = observations[guesses == i].mean(axis=0)    _inertia = distance_fct(old_centroids,centroids)    plot_centroids(centroids,old_centroids)    return _inertiaplot_clusters()while True:    step_one()    plot_clusters()    inertia = step_two()    print inertia,threshold    """    3. If the change in centroid locations is less than some threshold, stop    """    if inertia > threshold:        continue    else:        break
```

Nagging Questions-----------------So we found our clusters! Hey, that's great.  ...but wait: we cheated.We initialized the centroids randomly, sure, but we also decided to start with fivecentroids because we knew that there really *were* 5 artificially generatedgroups of data. Obviously, in the real world, we won't often be tasked with describingdata that we also happened to have generated.We made two questionable choices. First:> a. the number of classes that we have (call this K)We set K to 5. Go back up and assign a different value to K.  We'll still converge on a solution, but if we set K too low or high, we probably won't like what we find.> b. our initial stab at where those classes' centroids are locatedTry generating random data over and over and sooner or later you'll get some clustersthat are pretty hard to distinguish because they overlap so much. If you get a setof data where a bunch of observations all crowd together, then our approach of randomlyassigning initial centroid values will show some bad behavior. The centroids won't beable to "fan out" across the data the way we might like, and they will end up making divisions across groups of observations that intuitively seem like a natural whole.So how do we make the "right" decisions in the initialization step? How do we decidehow many clusters there are, and where they should start?As was mentioned before, there's quite a bit of research on this. like many problems inmachine learning, however, the solutions break out into two groups: ### 1. Search the entire space ###* do all the things! every combination of possible values for each parameter* a (somewhat unsatisfying) answer* it's often the only optionThe key, if you're going to try this a bajillion times, is that you need a way to compare the results across trials. If you happen to know the truth about class/clustermembership, this can be easy (see purity ratings, Rand index, V-measure, mutual information). But this is *unsupervised*, isn't it? There are basically two approaches,if we lack "ground truth" about the data:* within-class vs. across-class variance     * basically, a ratio of how tightly grouped members of the same cluster are, relative to their distance from members of other classes    * would let us diagnose cases where there are artificial splits over things thatseem to belong together* silhouette coefficient    * how close was each given point to "accidentally" falling into the next nearest cluster?    * given a sample x:>> $x_{within}$ = mean distance between x and its cluster-mates>> $x_{across}$ = mean distance between x and members of next-nearest cluster>>>> $s = \dfrac{(x_{across}) - (x_{within})}{max(x_{across},x_{within})}$ ### 2. Get Clever ###There have been a few other attempts at solving the problem of initialization in K-means. None of them help you pick K, but they might help by reducing the number oftrials you devote to any given K by initializing the centroids in a "smart" way. Here'sa sample:* Forgy: randomly choose k observations from the data set (we did this)* Random Partition: randomly assign a class to each observation* Kmeans++    0. choose one centroid at random from among observations    1. for each data point, compute $D(x)$: the distance between it and the nearest center:     2. choose a new data point at random, using weighted probability distribution where a point x is chosen with probability proportional to $D(x)^2$    3. repeat 1 and 2 until you've chosen K centers
