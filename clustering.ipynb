K-Means Clustering
==================
Let's say we have data that looks like the following: we have 1000 observations
that (unknown to us) are random points belonging to five different "classes." One 
way to think about this is that each datum is generated by a normal distribution 
around one of five means. The observations of these data each have an x and a y 
attribute.

```python
observations, real_labels = make_blobs(1000,centers=5, cluster_std=0.7)
plt = scatter(observations[:,0], observations[:,1]) 
```

Let's add some color to show which observations belong to each class:

```python
colors = ['red','green','blue','yellow','purple']
plt = scatter(observations[:,0], observations[:,1], 
              color=map(lambda x: colors[x],real_labels))
```

Our mission, then, is to try to derive these groupings using only the attributes
that we've observed (x and y) for each datum.  

Minimizing Inertia
------------------
What K-means does is try to find points in our attribute space (here, a 2-dimensional
space) that can be thought of as the "centers of gravity" (we'll call them centroids) 
for the observations we've collected. The assumption is that observations that are 
members of the same class will gravitate towards each other more than they would towards
members of some other class. If that's a reasonable assumption, then we can use K-means 
to find the most representative members of each class, and expect all of its fellow
class-members (classmates?) to crowd around it.

We're going to start by "guessing" at the number of classes that we have, and then do 
our best to divide the observations up so that each observation is a member of just one
class. The goal is for each class to have a stable center of gravity that its points 
crowd around.

If "guessing" in step one strikes you as odd, you're not crazy. The initialization step
of K-Means has an enormous effect on its eventual solution, and there's a whole body of
research into how to best accomplish this. Remember that we're guessing at two things:

> a. the number of classes that we have (call this K)
>
> b. our initial stab at where those classes' centroids are located

In and of itself, (b) seems like a crazy thing to do. That's okay, though: we're going
to revise our guess (based on actual evidence) often. Guessing at (a), though, is 
trickier. For demonstration purposes, we'll pick K=5, so we can see K-means succeed. 
Feel free to try some other values and see how they pan out, though. In practice, we
would try lots of differnt values for K and decide which one gives us the best 
model of the data.

Okay, so we start by randomly picking K points in that space (one for each class):

```python
import numpy as np
scatter(observations[:,0], observations[:,1],color='gray') 
hold(True)

# Initialization: Randomly select 5 observations as our centroids
K = 5
centroids = np.random.permutation(observations)[:K]
scatter(centroids[:,0],centroids[:,1],color='black',marker='x',s=100)
centroids
```

and then iterate these steps:

> 1. Assign each observation to the class of centroid closest to it
> 2. Take the mean of each class, decalre it the new class centroid
> 3. Measure the distance that between the old centroid and the new one. 
>> IF the distance is lower than some threshold, stop.
>>
>> ELSE continue

```python
guesses = np.zeros_like(real_labels)
threshold = 0.1
inertia = 100

def distance_fct(a,b):
    return np.linalg.norm(a - b)

def plot_clusters():
    """
    Just plotting some clusters. Ignore me.
    """
    scatter(observations[:,0], observations[:,1], 
            color=map(lambda x: colors[x],guesses))
    hold(True)

def plot_centroids(centroids,old_centroids):
    """
    Just plotting some centroids. Ignore me.
    """
    scatter(centroids[:,0], centroids[:,1], 
            color='black',s=100,marker='x')
    hold(True)
    plot([centroids[:,0],old_centroids[:,0]], 
         [centroids[:,1],old_centroids[:,1]],
         color='black')
    
def step_one():
    """ 
    1. Assign each observation to the class of the centroid closest to it
    """
    for id,observation in enumerate(observations):
        distances = [distance_fct(observation,centroid) for centroid in centroids]
        distances = np.array(distances)
        guesses[id] = distances.argmin()
    plot_clusters()
    
def step_two():
    """
    2. Take the mean of each class. Assign it as the new centroid for that 
       class
    """
    old_centroids = centroids.copy()
    for i in range(len(old_centroids)):
        centroids[i] = observations[guesses == i].mean(axis=0)
    _inertia = distance_fct(old_centroids,centroids)
    plot_centroids(centroids,old_centroids)
    return _inertia

plot_clusters()
while True:
    step_one()
    plot_clusters()
    inertia = step_two()
    print inertia,threshold
    """
    3. If the change in centroid locations is less than some threshold, stop
    """
    if inertia > threshold:
        continue
    else:
        break
```

*Note: the above is a hand-written, super dumb implementation of K-means. Not for use in production!*

Nagging Questions
-----------------
So we found our clusters! Hey, that's great.  

...but wait: we cheated.

We initialized the centroids randomly, sure, but we also decided to start with five
centroids because we knew that there really *were* 5 artificially generated
groups of data. Obviously, in the real world, we won't often be tasked with describing
data that we also happened to have generated.

We made two questionable choices. First:

> a. the number of classes that we have (call this K)

We set K to 5. Go back up and assign a different value to K.  We'll still converge 
on a solution, but if we set K too low or high, we probably won't like what we find.

> b. our initial stab at where those classes' centroids are located

Try generating random data over and over and sooner or later you'll get some clusters
that are pretty hard to distinguish because they overlap so much. If you get a set
of data where a bunch of observations all crowd together, then our approach of randomly
assigning initial centroid values will show some bad behavior. The centroids won't be
able to "fan out" across the data the way we might like, and they will end up making 
divisions across groups of observations that intuitively seem like a natural whole.

So how do we make the "right" decisions in the initialization step? How do we decide
how many clusters there are, and where they should start?

As was mentioned before, there's quite a bit of research on this. like many problems in
machine learning, however, the solutions break out into two groups: 

### 1. Search the entire space ###
* do all the things! every combination of possible values for each parameter
* a (somewhat unsatisfying) answer
* it's often the only option

The key, if you're going to try this a bajillion times, is that you need a way to 
compare the results across trials. If you happen to know the truth about class/cluster
membership, this can be easy (see purity ratings, Rand index, V-measure, mutual 
information). But this is *unsupervised*, isn't it? There are basically two approaches,
if we lack "ground truth" about the data:

* within-class vs. across-class variance 
    * basically, a ratio of how tightly grouped members of the same cluster are, 
relative to their distance from members of other classes
    * would let us diagnose cases where there are artificial splits over things that
seem to belong together
* silhouette coefficient
    * how close was each given point to "accidentally" falling into the next nearest 
cluster?
    * given a sample x:

>> $x_{within}$ = mean distance between x and its cluster-mates

>> $x_{across}$ = mean distance between x and members of next-nearest cluster

>>>> $s = \dfrac{(x_{across}) - (x_{within})}{max(x_{across},x_{within})}$ 

### 2. Go to Vegas! ###
* searching the entire space of parameter values is a waste of time 
    * the bookkeeping alone takes time
    * many changes will not shift the ultimate solution very much
* where initial cluster values are concerned, you don't have to be precise (shotgun, not sniper rifle)
* Instead of trying everything, for all values of k:
    * initialize with random starts
    * iterate thousands of times, take best solution(s)

### 3. Get Clever ###
There have been a few other attempts at solving the problem of initialization in 
K-means. None of them help you pick K, but they might help by reducing the number of
trials you devote to any given K by initializing the centroids in a "smart" way. Here's
a sample:

* Forgy: randomly choose k observations from the data set (we did this)
* Random Partition: randomly assign a class to each observation
* Kmeans++
    0. choose one centroid at random from among observations
    1. for each data point, compute $D(x)$: the distance between it and the nearest center: 
    2. choose a new data point at random, using weighted probability distribution 
where a point x is chosen with probability proportional to $D(x)^2$
    3. repeat 1 and 2 until you've chosen K centers

```python
%reset

# Doing K-Means in scikit-learn (http://scikit-learn.org/stable/auto_examples/cluster/plot_mini_batch_kmeans.html#example-cluster-plot-mini-batch-kmeans-py)

import time

import numpy as np
import pylab as pl

from sklearn.cluster import MiniBatchKMeans, KMeans
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.datasets.samples_generator import make_blobs

##############################################################################
# Generate sample data
np.random.seed(0)

batch_size = 45
centers = [[1, 1], [-1, -1], [1, -1]]
n_clusters = len(centers)
X, labels_true = make_blobs(n_samples=3000, centers=centers, cluster_std=0.7)

##############################################################################
# Compute clustering with Means

k_means = KMeans(init='k-means++', n_clusters=3, n_init=10)
t0 = time.time()
k_means.fit(X)
t_batch = time.time() - t0
k_means_labels = k_means.labels_
k_means_cluster_centers = k_means.cluster_centers_
k_means_labels_unique = np.unique(k_means_labels)

##############################################################################
# Compute clustering with MiniBatchKMeans

mbk = MiniBatchKMeans(init='k-means++', n_clusters=3, batch_size=batch_size,
                      n_init=10, max_no_improvement=10, verbose=0)
t0 = time.time()
mbk.fit(X)
t_mini_batch = time.time() - t0
mbk_means_labels = mbk.labels_
mbk_means_cluster_centers = mbk.cluster_centers_
mbk_means_labels_unique = np.unique(mbk_means_labels)

##############################################################################
# Plot result

fig = pl.figure(figsize=(8, 3))
fig.subplots_adjust(left=0.02, right=0.98, bottom=0.05, top=0.9)
colors = ['#4EACC5', '#FF9C34', '#4E9A06']

# We want to have the same colors for the same cluster from the
# MiniBatchKMeans and the KMeans algorithm. Let's pair the cluster centers per
# closest one.

distance = euclidean_distances(k_means_cluster_centers,
                               mbk_means_cluster_centers,
                               squared=True)
order = distance.argmin(axis=1)

# Plot KMeans
ax = fig.add_subplot(1, 3, 1)
for k, col in zip(range(n_clusters), colors):
    my_members = k_means_labels == k
    cluster_center = k_means_cluster_centers[k]
    ax.plot(X[my_members, 0], X[my_members, 1], 'w',
            markerfacecolor=col, marker='.')
    ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col,
            markeredgecolor='k', markersize=6)
ax.set_title('KMeans')
ax.set_xticks(())
ax.set_yticks(())
pl.text(-3.5, 1.8,  'train time: %.2fs\ninertia: %f' % (
    t_batch, k_means.inertia_))

# Plot MiniBatchKMeans
ax = fig.add_subplot(1, 3, 2)
for k, col in zip(range(n_clusters), colors):
    my_members = mbk_means_labels == order[k]
    cluster_center = mbk_means_cluster_centers[order[k]]
    ax.plot(X[my_members, 0], X[my_members, 1], 'w',
            markerfacecolor=col, marker='.')
    ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col,
            markeredgecolor='k', markersize=6)
ax.set_title('MiniBatchKMeans')
ax.set_xticks(())
ax.set_yticks(())
pl.text(-3.5, 1.8, 'train time: %.2fs\ninertia: %f' %
        (t_mini_batch, mbk.inertia_))

# Initialise the different array to all False
different = (mbk_means_labels == 4)
ax = fig.add_subplot(1, 3, 3)

for l in range(n_clusters):
    different += ((k_means_labels == k) != (mbk_means_labels == order[k]))

identic = np.logical_not(different)
ax.plot(X[identic, 0], X[identic, 1], 'w',
        markerfacecolor='#bbbbbb', marker='.')
ax.plot(X[different, 0], X[different, 1], 'w',
        markerfacecolor='m', marker='.')
ax.set_title('Difference')
ax.set_xticks(())
ax.set_yticks(())

pl.show()
```

K-Means: Summary
----------------
We've seen the effectiveness of K-Means clustering in finding the "centers of gravity" in our data and allowing us to assign each datapoint to one of those centroids. It can be an efficient way of discovering clusters, but we also noted some weaknesses:

* picking the right number of clusters is very important
* initializing the values of the centroids can have big impacts on the outcome.

To address these problems, we saw that it is necessary to to perform many clusterings with different parameter values in order to be confident in our findings. To help us, there are a few tools for clustering evaluation that we can look to.

Model-based EM Clustering
=========================
Introduction
------------
Above we described K-means as a process that tries to identify the most representative members of each class (the centroids) and then describe the rest of the observations in terms of their distance from those exemplars. 

In fact, K-means can be thought of in another way (without changing how the math actually works). We *could* think of it as discovering the underlying model that generated our data in the first place. Actually, in the example above this is a pretty easy way to think about it, since the data was, in fact, generated by an underlying model. 

So imagine K-means as discovering hidden "emitters" of random data. Think of the blobs as the result of little data generators, located at different points in a 2D space, that spit out data that spreads around them in a random, uniform way. It's sortof like looking at a bunch of puddles gathering at the bottom of a boat: the first thing you want to know is where the holes are! This gets at another assumption of K-means that we haven't touched on: the clusters in K-means are assumed to be "spheres," meaning that they have a center (their mean) and that the members of the class can be well described using the normal Gaussian distribution, with a standard deviation. As we'll see, that's not a good assumption for many other kinds of data, whose clusters are not well-described using the normal distribution.

Thought of this way, K-means is actually just a special case of Model-based clustering. In model-based clustering, instead of trying to find representative values of the data we see, we try to uncover the model that generated that data. K-means is an efficient way of doing this when the right assumptions apply, but having a more general procedure for Model-based clustering is immensely powerful. It lets us follow a generalized version of the steps in the K-means algorithm, while freeing us up to consider models that are not described using the normal distribution.  

Some examples of distributions that have been found particularly suitable to describing real-world events are:

* **Bernoulli distribution:** discrete events with binary outcomes (clinical trial outcomes, coin tosses, presence/abscence of a word in a document)
* **Poisson distribution:** frequency of an event over a span of time (bus schedules, retail checkout lines, frequency of a word in a document, deaths due to being kicked in the head by a horse)
* **Beta distribution:** probability of a probability, given a likely range (allele frequencies in a population, time spent on tasks in a project, sunshine data, batting averages)
* **Power law distribution:** frequency of an event varies as a power of one of its attributes (word frequency counts, sizes of power outages, frequency of digits in accounting ledgers, populations of cities, lurking in virtual communities, sizes of earthquakes/moon craters/solar flares)

*RealFact: these descriptions are ridiculously oversimplified and I picked examples that I thought were interesting or funny. Especially flimsy: the power law examples*

The point is, if we care about things in the real world, we'll probably make some sets of observations that aren't clusterable using K-means. Luckily, there's a way to generalize the best parts of the algorithm while leaving room to consider alternative distributions.

Maximum Likelihood
------------------
Let's see what it takes to really think of K-means as an exercise in uncovering an underlying model. Model-based clustering is always seeking to find the model that maximizes the probability of producing the data that we're looking at. In math-speak, we want to set the parameters (here, the number of clusters and their locations) that maximizes the likelihood of the data. That might seem a bit backward, but it works! 

The **maximum likelihood criterion** is the law of the land in model-based clustering. We're always going to be messing with $\Theta$ (the parameters) until we're confident that we've maxmized 

>>> $\displaystyle L(D|\Theta)$

...or the likelihood of our data, $D$, given $\Theta$. This likelihood function can be any function that tells us how good we did at clustering. From here on, I'll introduce more mathy descriptions, but do my best to translate each one into a kind of self-documenting, Python-flavored pseudocode that will hopefully make it easier to understand. For instance, $\Theta$ is going to refer to a collection of different settings for our parameters. How many? Well, we don't know, so let's just say we have $J$ parameter-setting objects.

*Greek:*

>>> $\displaystyle\Theta = {\Theta_1,\Theta_2,...,\Theta_J}$

*English:*

    all_param_val_sets = [{param_a:val_a_1, param_b:val_b_1}, 
                          {param_a:val_a_2, param_b:val_b_2}, 
                              ... 
                          {param_a:val_a_J, param_b:val_b_J}]    

In what we called "step 2" of K-means, we always moved the centroids to the mean of their (newly-constructed) clusters. What we were doing was minimizing the overall residual sum of squares (RSS) of the cluster, which is: 

*Greek:*

>>> $\displaystyle\sqrt{\sum\limits_{x \in \omega}\mid x - \mu(\omega)\mid}$ 

*English:*

    def compute_rss():
        sum_diffs = 0
        for x in cluster:
            sum_diffs += x - mean(cluster)
        return sqrt(sum_diffs)

of the sum of the distances of the cluster members from their centroid. If you assume your clusters are, in reality, normally distributed, then by definition the way to reduce RSS is to move the centroid to the mean. That means that the minimal value of RSS *is* the maximum likelihood criteria for K-means. 

So however we define it, `minimize_rss(params)` is the objective function that tells us how good our clusters are. If it returns a likelihood measure, then we can say that we get our best values for $\Theta$ by computing:

*Greek:*

>>> $\displaystyle\DeclareMathOperator*{\argmax}{argmax} \Theta = \argmax\limits_{\Theta}\;L(D|\Theta)$

*English:*

    likelihoods = [0] * len(all_param_val_sets)
    
    for i,param_val_set in enumerate(all_param_val_sets):
        likelihoods[i] = minimize_rss(param_val_set,data)
    
    maximum_likelihood = max(likelihoods)
    best_param_val_set = all_param_val_sets.index(maximum_likelihood)

K-means works well because, for a given set of centroids, $\displaystyle e^{-RSS}$  is proportional to the likelihood of that model. The pseudocode above is reliant on minimize_rss(), which works well for K-means. 

This gives us the framework, though, to extend this process to datasets that K-means can't handle. All we'd need is an appropriate function to swap in for `minimize_rss()`, and otherwise recreate the same steps. 

The general form of the maximum likelihood criterion is a little more specific about $L(D|\Theta)$, describing it in terms of probability functions. It remains agnostic, however, to the distributions that those probabilities are based on:

*Greek:*

>>> $\displaystyle\DeclareMathOperator*{\argmax}{argmax} \Theta = \argmax\limits_{\Theta}\;L(D|\Theta) = \argmax\limits_{\Theta}\; log\prod_{n=1}^NP(x_n|\Theta) = \argmax\limits_{\Theta}\; \sum_{n=1}^{N}log P(x_n|\Theta)$

*English:*

    all_param_val_sets = [{param_1:val_a, param_2:val_b}, {param_1:val_c, param_2:val_d}, ...]
    log_likelihoods = [0] * len(all_param_val_sets)
    
    for i,param_val_set in enumerate(all_param_val_sets):
        log_likelihood = 0
        for x in data:
            log_likelihood += log(probability(x,param_val_set))
        log_likelihoods[i] = log_likelihood
    
    maximum_log_likelihood = max(log_likelihoods)
    best_param_val_set = all_param_val_sets.index(maximum_log_likelihood)

Okay, cool! Now we have something that's modular enough to use with any old probability distribution. Once we decide on one, we can compare lots of different configurations of our parameters and decide which one gives us the best chance of constructing a useful model of our data. 

Still, it's going to be tough to search through all possible values of all of our parameters. They might have big, maybe continuous ranges, and there might be a lot of them. 

Undecided Again, Naturally
--------------------------
When we're trying to model the underlying structure of our data, we'd prefer to choose a model and then attempt to find the parameters of that model that have the highest log-likelihood of generating the data. Our problem when clustering naively, however, is not only that we don't know what those parameter values should be, but we also have no idea what the right model actually looks like, and the set of possibilities is large or infinite. 

The general problem of estimating the best values for the parameters of any model-based clustering is summed up in two circular statements (Manning and Schutze, 2003):

>>> a. if we knew the value of our parameters, we could compute the expected values of the hidden structure of the model.

>>> b. if we knew the expected values of the hidden structure of the model, then we could predict the value of our parameters that gives the maximum likelihood of having generated our data.

Helpful, right? Remember back in school when things were easy and all you had to do was find the best line to fit to some bunch of points? Weren't those happy times?

Ordinary least squares regression is solvable because we make a decision about the model itself. What we need to solve for is just the optimal values of its parameters, which can be done analytically with some simple calculus.

In the case of model-based clustering, though, both the parameters of the model and the structure of the model itself are in flux. The only way to solve for the parameters is a set of nonlinear equations for which no analytical methods of solution are known.

The olllld "Texas Two Step"
---------------------------
*RealFact: no one has ever called EM "the ollld 'Texas Two Step'"*

The Expectation-Maximization algorithm is an iterative solution to the circular statments (a) and (b). The circularity at the initialization step, which makes a guess at $\Theta$, after which we alternate between two steps of estimation and revision. Let's look again at K-means to see what those steps are like.

In K-means, we had three steps that we iterated over in order to discover the model underlying the data (ie, our centroid placements). Let's attach some new names to them:

1. **Expectation:** Assign each observation to the class of centroid closest to it
2. **Maximization:** Take the mean of each class, decalre it the new class centroid
3. **Stopping Criterion:** Measure the distance that between the old centroid and the new one.

The *Expectation* (E) step involves assigning observations to the cluster they seem to "belong" to. In K-means, this is the  assignment of an observation to some cluster based on a distance metric that tells us which centroid is closest to it. 

The *Maximization* (M) step involves re-estimating the parameters to achieve the maximum likelihood of the data given the assignments as they are at the end of the E step. In K-means, since the likelihood is proportional to the RSS, we alter our parameters (the values of the centroids) so that each centroid is now at the mean of its class members' values.

The *Stopping Criterion* isn't really a step. It's a calculation of the marginal improvement of our model after this revision. It's provable that these steps are **monotonic**, meaning that our model is getting better all the time. What (3) is really asking is how much we've learned after E and M. Put another way, it's asking how much our parameters have changed to give us our latest improvement. If it's not very much, we say that we **converged** on an optimal setting of those parameters. In K-means, we could do this easily by asking how far the centroids had moved, ie, how much of a change in our model we had to make in order to increase the probability of its having generated this data. Again, if the answer is "not much," then we say that we've converged, ie, this model is "good enough."

Why complicate such a straightforward algorithm? To repeat, this more generalized procedure gives us the freedom to cluster sets for which the assumptions of K-means don't hold (spherical, normal clusters). As we'll see, there are also other advantages to tackling the clustering problem in this way, and we'll learn more about each individual observation.

Mixture Models
--------------
We can now proceed with model-based clustering using any distribution as our model type. That means that we assume the observations we made were produced by several hidden models, each of which is a mulivariate distribution (multivariate because each attribute will have its own likelihood for possible values). 

So what we'll end up with to describe the data is a *mixture model*, which is a probabilistic model that can describe all observations as the product of multiple *components*, each with their own parameter settings. A simplistic version involves a one-dimensional dataset that can be modeled as the product of two underlying normal distributions:

```python
#Adapted from figure 16.9 in Witten & Frank (2005)
import matplotlib.mlab as mlab

x = np.linspace(30,80,1000)
class_a = np.random.normal(50,5,60) # class A is generated with probability 0.6
class_b = np.random.normal(65,2,40) # class B is generated with probability 0.4

hold(True)
# The data set, colored according to the component that generated it
scatter(class_a,zeros_like(class_a),color="blue")
scatter(class_b,zeros_like(class_b),color="green")

# The probability distibutions of the two components needed to make up a mixture model 
# that adequately describes the distribution of the data.
plot(x,mlab.normpdf(x,50,5)) 
plot(x,mlab.normpdf(x,65,2))
```

This two-component Gaussian mixture model is a little easier to grasp intuitively than what we're going to be doing later, so it's worth understanding how EM will help us, here. Imagine only seeing the data scattered at the bottom. Suppose we made a good guess and decided that data could be represented by two underlying Gaussian models. What would we have to learn to describe those mystery models? 

* their shape 
    * for Gaussians, described in terms of mean and variance
    * four parameters: $\mu_A, \sigma_A, \mu_B, \sigma_B
* their weights 
    * for some random observation, how likely is it to have been generated by component A? 
    * one more parameter: $p(A)$ (after all, $p(B)$ will just be $1 - p(A)$)

So that's five parameters we'd need to learn. Not easy, but we can always count on our good buddy EM to help us find them quickly. The pattern is, again, familiar. 

**parameter initialization:** 

* We make initial guesses for the five parameters. 
* Because we're dealing with a mixture of components, we also want to take into account our belief in how likely a given component is to generate a given observation, no matter what its value. Basically we want to know how much wieght each component carried in the generation of the data. We start by assuming that this is equal for all classes: 

>> $\pi_A = \pi_B = \frac{1}{number\;of\;components} = \frac{1}{2} = 0.50$

Then, repeat until convergence:

**Expectation:** For each observation, assign it probabilities of having been generated by each class. 

* This is easy, with Gaussians, since all we need to know is the observation's distance from the mean and the variance of the distribution (both of which we choose the first time around).
* As a result of this, each observation $x$ will have a probability $p(x|C)$ of being a member of each class C

**Maximization:** For each cluster, revise the two parameters $\mu$ and $\sigma$ to maximize the likelihood of the data given the current weights of assignment.

* We use the probabilities computed in the E step to perform a weighted average for each cluster. This gives us the new $\mu$ for that cluster.

>> $\mu_A = \frac{w_{A1}x_1 + w_{A2}x_2 + ... + w_{An}x_n}{w_{A1} + w_{A2} + ... + w_{An}}\;\;$  where $w_{Ai}$ is $p(x_i|A)$

>> $\mu_B = \frac{w_{B1}x_1 + w_{B2}x_2 + ... + w_{Bn}x_n}{w_{B1} + w_{B2} + ... + w_{Bn}}\;\;$  where $w_{Bi}$ is $p(x_i|B)$

* Then, we use a weighted average of the squared error to compute the new $\sigma$

>> $\sigma_A = \frac{wA_1(x_1-\mu_A)^2 + wA_2(x_2-\mu_A)^2 + ... + wA_n(x_n-\mu_A)^2}{wA_1 + wA_2 + ... + wA_n}\;\;$ where $wA_i$ is $p(x_i|A)$

>> $\sigma_B = \frac{wB_1(x_1-\mu_B)^2 + wB_2(x_2-\mu_B)^2 + ... + wB_n(x_n-\mu_B)^2}{wB_1 + wB_2 + ... + wB_n}\;\;$ where $wB_i$ is $p(x_i|B)$

* And again, because we're dealing with a mixture of components, we need to update our belief in *their* relative weights, as well:

>> $\pi_A = \frac{w_1 + w_2 + ... + w_n}{w_1 + w_2 + ... + w_n}\;\;$ where $w_i$ is $p(x_i|A)$

```python
# Gaussian EM in Python, from http://scikit-learn.org/stable/auto_examples/mixture/plot_gmm_pdf.html#example-mixture-plot-gmm-pdf-py
import numpy as np
import pylab as pl
from sklearn import mixture

n_samples = 300

# generate random sample, two components
np.random.seed(0)
C = np.array([[0., -0.7], [3.5, .7]])
X_train = np.r_[np.dot(np.random.randn(n_samples, 2), C),
                np.random.randn(n_samples, 2) + np.array([20, 20])]

clf = mixture.GMM(n_components=2, covariance_type='full')
clf.fit(X_train)

x = np.linspace(-20.0, 30.0)
y = np.linspace(-20.0, 40.0)
X, Y = np.meshgrid(x, y)
XX = np.c_[X.ravel(), Y.ravel()]
Z = np.log(-clf.eval(XX)[0])
Z = Z.reshape(X.shape)

CS = pl.contour(X, Y, Z)
CB = pl.colorbar(CS, shrink=0.8, extend='both')
pl.scatter(X_train[:, 0], X_train[:, 1], .8)

pl.axis('tight')
pl.show()
```

Soft Assignment
---------------
Maybe the most interesting difference between K-means and EM is the different attitudes they take toward class membership. K-means performs **hard assignment** of observations to their potential (and ultimate) classes. In other words, if you're a member of class A, you can't also be a member of class B or any other class. 

EM, on the other hand, performs **soft assignment**. For any given data point, we know how likely it is to be a member of each class. So although we'll summarily declare that it is a member of the component that maximizes its log-likelihood, we can also remember how confident we are in that statement, and possibly update that belief when we get new observations.

Documents as Term Vectors
-------------------------
When we look at document $d_i$, and want to describe it in terms of our total vocabulary (the 18 unique terms of corpus $D$), one useful way to do so is to simply state whether or not the word occurred in that document. This translates to a binary value: 1 if the term did occur, and 0 if the word did not. That gives us the following observations:

```python
# This is a really roundabout way of constructing a document x term matrix. 
# It's for demonstration purposes!

from collections import defaultdict
td_dict = defaultdict(lambda: defaultdict(int))

vocabulary = term_counts.keys()

for doc_id,text in texts.iteritems():
    for term in vocabulary:
        if term in text:
            td_dict[doc_id][term] = 1
        else:
            td_dict[doc_id][term] = 0
print "Raw Text"

print docs

print "\n\nDocument x Term Matrix"
td_frame = DataFrame.from_dict(td_dict,orient="index")
td_frame.columns = td_frame.columns.map(lambda x: '$t_{'+x+'}$')  # just dressing it up
td_frame.index = td_frame.index.map(lambda x: '$d_{'+str(x)+'}$') # just dressing it up
td_frame
```

The Bernoulli Distribution
--------------------------
Because we're looking at binary events for each of our attributes, it makes sense to assume that the data was produced by a mixture model whose components conform to the Bernoulli distribution. The Bernoulli distribution assumes that there are two possible mutually exclusive outcomes for every observation. 

The classic example is a coin toss, with no assumptions about whether the coin is fair. What are the odds, in a given toss, that a coin will land on Heads? Let's say that $H = 1$ if we do get a Heads result, and $H = 0$ if we don't. Then we can model the distribution of the probability of $\displaystyle P(H=x \in\;${$\;1,0\;$}$)$ using the variable $p$:

>>>> $\displaystyle P(H=x) = p^x(1-p)^{1-x}$

eg:

>>>> $\displaystyle P(H=1) = p^1(1-p)^{1-1} = p^1(1-p)^0 = p$

>>>> $\displaystyle P(H=0) = p^0(1-p)^{1-0} = p^0(1-p)^1 = 1-p$

or a small example from our data: word frequencies, the probability of seeing, say, $t_{ghana}$ in any given document $d|i} in our toy data:

>>>> $P(t_{ghana}=1|D) = {\frac{2}{11}}^{1}(1 - \frac{2}{11})^{1-1} = \frac{2}{11}$

>>>> $P(t_{ghana}=0|D) = {\frac{2}{11}}^{0}(1 - \frac{2}{11})^{1-0} = (1 - \frac{2}{11}) = \frac{9}{11}$

Obviously, there are different values for $p$ and $1-p$ for each term in our vocabulary.

Also, remember that the values we just calculated were in relation to the entire corpus $D$. We won't be doing that again, because for the sake of clustering, we want to assume that the documents we observed were generated by a number of different models, one for each cluster that we expect to find.

EM Clustering with a Mixture of Multivariate Bernoulli Distributions
--------------------------------------------------------------------
So now we're ready to perform EM to cluster these texts. We've decided to treat our corpus as the product of a model which is made up of multiple components, each one a multivariate Bernoulli distribution. Again, we'll be doing soft assignment, without having to definitely decide which component generated each document. Instead, each component is going to be seen as generating each document, but with different probabilities.

This is getting a little complex, so let's stop to appreciate the different levels we're dealing with. We're going to assume:

* our data was created by a "mixture model" with K components

>> $\Theta = $ { $\Theta_1,...,\Theta_K$ }

* each component has a likelihood, which is the likelihood a given document $d_n$ is in cluster $c_k$ if we have no information about $d_n$ (sometimes called a "prior")

>> $\pi_k = P(\Theta_k)$

* and set of parameter values (one for each term in our vocabulary)

>> $\Theta_k = (\theta_{1k}, ... , \theta_{Mk})$

* each parameter is expressed as the probability that a term is present in a document given the document's class

>> $q_{mk} = P(t_m = 1 | c_k) = p_{mk}^1(p_{mk} - 1)^{1 - 1} = p_{mk}$

Got it? Just like above, we model each attribute (here, term presence/absence) according to our chosen distribution (the Bernoulli distribution) and we try to find the model of clusters C with parameters $\Theta$ that maximize the likelihood of the data we've observed.
